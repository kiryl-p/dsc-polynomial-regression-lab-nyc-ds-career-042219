## Create the best plot using polynomials!

Here is a plot with a clearly non-linear shape. Begin to think about what degree polynomial you believe will fit it best.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

def pol(x):
    return x * np.cos(x)

x = np.linspace(0, 12, 100)
rng = np.random.RandomState(1234)
rng.shuffle(x)
x = np.sort(x[:25])
y = pol(x) + np.random.randn(25)*2

plt.scatter(x, y, color='green', s=50, marker='.')

plt.show();
```

Now it's time to determine the optimal degree of polynomial features for a model that is fit to this data

## Solution

```python
## your code here
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly_2= PolynomialFeatures(2, interaction_only= False)
poly_3= PolynomialFeatures(3)
poly_4= PolynomialFeatures(4)
```

```python
poly_5= PolynomialFeatures(5)
x_fin5= poly_5.fit_transform(x)
```

```python
x= pd.DataFrame(x)
```

```python
x_fin2 = poly_2.fit_transform(x)
x_fin3 = poly_3.fit_transform(x)
x_fin4 = poly_4.fit_transform(x)
```

```python
print('The transformed feature names are: {}'.format(poly_2.get_feature_names()))
print('------------------')
print('The first row of transformed data is: {}'.format(x_fin2[0]))
```

```python
x_linspace = pd.DataFrame(np.linspace(0,12,50))


reg_poly2 = LinearRegression().fit(x_fin2, y)
predict_x = reg_poly2.predict(x_fin2)
x_linspace_fin2 = poly_2.fit_transform(x_linspace)
y_poly2_pred = reg_poly2.predict(x_linspace_fin2)

reg_poly3 = LinearRegression().fit(x_fin3, y)
predict_x = reg_poly3.predict(x_fin3)
x_linspace_fin3 = poly_3.fit_transform(x_linspace)
y_poly3_pred = reg_poly3.predict(x_linspace_fin3)

reg_poly4 = LinearRegression().fit(x_fin4, y)
predict_x = reg_poly4.predict(x_fin4)
x_linspace_fin4 = poly_4.fit_transform(x_linspace)
y_poly4_pred = reg_poly4.predict(x_linspace_fin4)

reg_poly5 = LinearRegression().fit(x_fin5, y)
predict_x = reg_poly5.predict(x_fin5)
x_linspace_fin5 = poly_5.fit_transform(x_linspace)
y_poly5_pred = reg_poly5.predict(x_linspace_fin5)




plt.figure(figsize= (8,8))
plt.scatter(x, y, color='green', s=50, marker='o')
plt.plot(x_linspace,y_poly2_pred)
plt.plot(x_linspace,y_poly3_pred)
plt.plot(x_linspace,y_poly4_pred)
plt.plot(x_linspace,y_poly5_pred, color='m', linewidth= 3)


plt.show();
```
